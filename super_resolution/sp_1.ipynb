```python
import numpy as np
import math
import matplotlib.pyplot as plt
from tensorflow.keras import datasets

# PSNR (입력 값 범위 [0,1] 가정)
def psnr(pred: np.ndarray, target: np.ndarray, eps: float = 1e-10) -> float:
    mse = np.mean((pred - target) ** 2)
    if mse < eps:  # 완전 동일할 때
        return 99.0
    return 10.0 * math.log10(1.0 / mse)
```

```python
# Fashion-MNIST 다운로드 & 로드
(train_images, _), (test_images, _) = datasets.fashion_mnist.load_data()

# [0,1] 정규화 (HR = 원본 28x28)
train_HR = train_images.astype(np.float32) / 255.0 
test_HR  = test_images.astype(np.float32)  / 255.0 

print("Train HR:", train_HR.shape, " Test HR:", test_HR.shape)

# 기본: 픽셀 스킵(::2)로 14x14 LR 생성
# train_LR = train_HR[:, ::2, ::2]        
# test_LR  = test_HR[:,  ::2, ::2]        

# area 평균으로 LR 생성
train_LR = np.stack([area_down2(x) for x in train_HR], axis=0)
test_LR  = np.stack([area_down2(x) for x in test_HR],  axis=0)

print("Train LR:", train_LR.shape, " Test LR:", test_LR.shape)
```

```python
# Bilinear 보간 함수
def bilinear_x2(img14: np.ndarray) -> np.ndarray:
    h, w = img14.shape
    H, W = h * 2, w * 2
    out = np.empty((H, W), dtype=np.float32)

    for y in range(H):
        v = (y + 0.5) / 2.0 - 0.5
        v0 = int(np.floor(v)); v1 = min(v0 + 1, h - 1); v0 = max(v0, 0)
        beta = v - v0
        for x in range(W):
            u = (x + 0.5) / 2.0 - 0.5
            u0 = int(np.floor(u)); u1 = min(u0 + 1, w - 1); u0 = max(u0, 0)
            alpha = u - u0

            top = (1 - alpha) * img14[v0, u0] + alpha * img14[v0, u1]
            bot = (1 - alpha) * img14[v1, u0] + alpha * img14[v1, u1]
            out[y, x] = (1 - beta) * top + beta * bot

    return np.clip(out, 0.0, 1.0)

# 테스트
idx = 0
HR = test_HR[idx]
LR = test_LR[idx]
SR_lin = bilinear_x2(LR)
p_lin = psnr(SR_lin, HR)

# 시각화
fig, ax = plt.subplots(1,3, figsize=(9,3))
ax[0].imshow(LR, cmap='gray', vmin=0, vmax=1); ax[0].set_title("LR 14×14"); ax[0].axis('off')
ax[1].imshow(HR, cmap='gray', vmin=0, vmax=1); ax[1].set_title("HR 28×28"); ax[1].axis('off')
ax[2].imshow(SR_lin, cmap='gray', vmin=0, vmax=1); ax[2].set_title(f"Linear\n{p_lin:.2f} dB"); ax[2].axis('off')
plt.tight_layout(); plt.show()
```

```python
# Keys cubic kernel (a=-0.5: Catmull-Rom)
def _cubic_kernel(t: float, a: float = -0.5) -> float:
    t = abs(t)
    if t <= 1:
        return (a + 2) * t**3 - (a + 3) * t**2 + 1
    if t < 2:
        return a * t**3 - 5*a * t**2 + 8*a * t - 4*a
    return 0.0

# Bicubic 보간 함수
def bicubic_x2(img14: np.ndarray, a: float = -0.5) -> np.ndarray:
    h, w = img14.shape
    H, W = h * 2, w * 2
    out = np.empty((H, W), dtype=np.float32)

    for y in range(H):
        v = (y + 0.5) / 2.0 - 0.5
        vi = int(np.floor(v)); dv = v - vi
        ys = [min(max(vi + j, 0), h - 1) for j in [-1, 0, 1, 2]]
        wy = np.array([_cubic_kernel(dv - j, a) for j in [-1, 0, 1, 2]], dtype=np.float32)

        for x in range(W):
            u = (x + 0.5) / 2.0 - 0.5
            ui = int(np.floor(u)); du = u - ui
            xs = [min(max(ui + i, 0), w - 1) for i in [-1, 0, 1, 2]]
            wx = np.array([_cubic_kernel(du - i, a) for i in [-1, 0, 1, 2]], dtype=np.float32)

            patch = img14[np.ix_(ys, xs)]
            out[y, x] = wy @ patch @ wx

    return np.clip(out, 0.0, 1.0)

# 테스트
idx = 0
HR = test_HR[idx]
LR = test_LR[idx]
SR_cub = bicubic_x2(LR)
p_cub = psnr(SR_cub, HR)

# 시각화
fig, ax = plt.subplots(1,3, figsize=(9,3))
ax[0].imshow(LR, cmap='gray', vmin=0, vmax=1); ax[0].set_title("LR 14×14"); ax[0].axis('off')
ax[1].imshow(HR, cmap='gray', vmin=0, vmax=1); ax[1].set_title("HR 28×28"); ax[1].axis('off')
ax[2].imshow(SR_cub, cmap='gray', vmin=0, vmax=1); ax[2].set_title(f"Cubic\n{p_cub:.2f} dB"); ax[2].axis('off')
plt.tight_layout(); plt.show()
```

```python
# --- nearest: 14x14 -> 28x28 ---
def nearest_x2(img14: np.ndarray) -> np.ndarray:
    h, w = img14.shape
    H, W = h * 2, w * 2
    out = np.empty((H, W), np.float32)

    for y in range(H):
        v = (y + 0.5) / 2.0 - 0.5
        v0 = int(np.round(v))
        v0 = min(max(v0, 0), h - 1)
        for x in range(W):
            u = (x + 0.5) / 2.0 - 0.5
            u0 = int(np.round(u))
            u0 = min(max(u0, 0), w - 1)
            out[y, x] = img14[v0, u0]

    return out

# --- 사용 예 (한 장 시각화) ---
idx = 0
HR = test_HR[idx]      # (28,28)
LR = test_LR[idx]      # (14,14)
SR_near = nearest_x2(LR)
p_near = psnr(SR_near, HR)

fig, ax = plt.subplots(1, 3, figsize=(9, 3))
ax[0].imshow(LR, cmap='gray', vmin=0, vmax=1); ax[0].set_title("LR 14×14"); ax[0].axis('off')
ax[1].imshow(HR, cmap='gray', vmin=0, vmax=1); ax[1].set_title("HR 28×28"); ax[1].axis('off')
ax[2].imshow(SR_near, cmap='gray', vmin=0, vmax=1); ax[2].set_title(f"Nearest\n{p_near:.2f} dB"); ax[2].axis('off')
plt.tight_layout(); plt.show()
```

```python
# --- lanczos kernel ---
def _lanczos_kernel(t: np.ndarray, a: int = 3) -> np.ndarray:
    t = np.asarray(t, np.float32)
    out = np.sinc(t) * np.sinc(t / a)       # sinc(x)=sin(pi x)/(pi x)
    out[np.abs(t) >= a] = 0.0               # 지지 범위 밖은 0
    return out

# --- lanczos: 14x14 -> 28x28 ---
def lanczos_x2(img14: np.ndarray, a: int = 3, normalize: bool = True) -> np.ndarray:
    """
    a=3 (Lanczos-3)이 보통의 기본값. a가 커지면 샤프하지만 ringing 위험↑
    normalize=True 이면 경계에서 가중치 합을 1로 정규화해 밝기 보존.
    """
    h, w = img14.shape
    H, W = h * 2, w * 2
    out = np.empty((H, W), np.float32)

    for y in range(H):
        v = (y + 0.5) / 2.0 - 0.5
        v0 = int(np.floor(v))
        js = np.arange(v0 - (a - 1), v0 + a + 1)   # 길이 2a+1
        js_clamped = np.clip(js, 0, h - 1)
        wy = _lanczos_kernel(v - js, a)

        for x in range(W):
            u = (x + 0.5) / 2.0 - 0.5
            u0 = int(np.floor(u))
            is_ = np.arange(u0 - (a - 1), u0 + a + 1)
            is_clamped = np.clip(is_, 0, w - 1)
            wx = _lanczos_kernel(u - is_, a)

            patch = img14[np.ix_(js_clamped, is_clamped)]  # (2a+1, 2a+1)
            # 경계에서 가중치 합이 1이 되도록 정규화(권장)
            if normalize:
                sw = wy.sum()
                if sw > 1e-8: wy = wy / sw
                sw = wx.sum()
                if sw > 1e-8: wx = wx / sw

            out[y, x] = wy @ patch @ wx

    return np.clip(out, 0.0, 1.0)

# --- 사용 예 (한 장 시각화) ---
idx = 0
HR = test_HR[idx]
LR = test_LR[idx]
SR_lanc = lanczos_x2(LR, a=3)
p_lanc = psnr(SR_lanc, HR)

fig, ax = plt.subplots(1, 3, figsize=(9, 3))
ax[0].imshow(LR, cmap='gray', vmin=0, vmax=1); ax[0].set_title("LR 14×14"); ax[0].axis('off')
ax[1].imshow(HR, cmap='gray', vmin=0, vmax=1); ax[1].set_title("HR 28×28"); ax[1].axis('off')
ax[2].imshow(SR_lanc, cmap='gray', vmin=0, vmax=1); ax[2].set_title(f"Lanczos-3\n{p_lanc:.2f} dB"); ax[2].axis('off')
plt.tight_layout(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets

# -----------------------------
# 0) 데이터 로드 (공통)
# -----------------------------
(_, _), (test_images, _) = datasets.fashion_mnist.load_data()
HR = test_images[0].astype(np.float32) / 255.0   # (28,28)
LR = HR[::2, ::2]                                # (14,14) : 픽셀 스킵 다운샘플

# -----------------------------
# 1) 보간 함수들 (half-pixel 정렬)
# -----------------------------
def nearest_x2(img14: np.ndarray) -> np.ndarray:
    h, w = img14.shape; H, W = h*2, w*2
    out = np.empty((H, W), np.float32)
    for y in range(H):
        v = (y + 0.5)/2.0 - 0.5
        v0 = int(np.round(v)); v0 = np.clip(v0, 0, h-1)
        for x in range(W):
            u = (x + 0.5)/2.0 - 0.5
            u0 = int(np.round(u)); u0 = np.clip(u0, 0, w-1)
            out[y, x] = img14[v0, u0]
    return out

def bilinear_x2(img14: np.ndarray) -> np.ndarray:
    h, w = img14.shape; H, W = h*2, w*2
    out = np.empty((H, W), np.float32)
    for y in range(H):
        v = (y + 0.5)/2.0 - 0.5
        v0 = int(np.floor(v)); v1 = min(v0+1, h-1); v0 = max(v0, 0)
        beta = v - v0
        for x in range(W):
            u = (x + 0.5)/2.0 - 0.5
            u0 = int(np.floor(u)); u1 = min(u0+1, w-1); u0 = max(u0, 0)
            alpha = u - u0
            top = (1-alpha)*img14[v0,u0] + alpha*img14[v0,u1]
            bot = (1-alpha)*img14[v1,u0] + alpha*img14[v1,u1]
            out[y, x] = (1-beta)*top + beta*bot
    return np.clip(out, 0.0, 1.0)

def _cubic_kernel(t: float, a: float = -0.5) -> float:
    t = abs(t)
    if t <= 1:  return (a+2)*t**3 - (a+3)*t**2 + 1
    if t < 2:   return a*t**3 - 5*a*t**2 + 8*a*t - 4*a
    return 0.0

def bicubic_x2(img14: np.ndarray, a: float = -0.5) -> np.ndarray:
    h, w = img14.shape; H, W = h*2, w*2
    out = np.empty((H, W), np.float32)
    for y in range(H):
        v = (y + 0.5)/2.0 - 0.5
        vi = int(np.floor(v)); dv = v - vi
        ys = [np.clip(vi+j, 0, h-1) for j in [-1,0,1,2]]
        wy = np.array([_cubic_kernel(dv-j, a) for j in [-1,0,1,2]], np.float32)
        sw = wy.sum();  wy = wy/sw if sw != 0 else wy  # 경계 안정화(선택)
        for x in range(W):
            u = (x + 0.5)/2.0 - 0.5
            ui = int(np.floor(u)); du = u - ui
            xs = [np.clip(ui+i, 0, w-1) for i in [-1,0,1,2]]
            wx = np.array([_cubic_kernel(du-i, a) for i in [-1,0,1,2]], np.float32)
            sw = wx.sum();  wx = wx/sw if sw != 0 else wx
            patch = img14[np.ix_(ys, xs)]
            out[y, x] = wy @ patch @ wx
    return np.clip(out, 0.0, 1.0)

def _lanczos_kernel_1d(t: np.ndarray, a: int = 3) -> np.ndarray:
    t = np.asarray(t, np.float32)
    out = np.sinc(t) * np.sinc(t / a)
    out[np.abs(t) >= a] = 0.0
    return out

def lanczos_x2(img14: np.ndarray, a: int = 3, normalize: bool = True) -> np.ndarray:
    h, w = img14.shape; H, W = h*2, w*2
    out = np.empty((H, W), np.float32)
    for y in range(H):
        v = (y + 0.5)/2.0 - 0.5
        v0 = int(np.floor(v))
        js = np.arange(v0-(a-1), v0+a+1); js_c = np.clip(js, 0, h-1)
        wy = _lanczos_kernel_1d(v - js, a)
        if normalize and wy.sum() != 0: wy = wy/wy.sum()
        for x in range(W):
            u = (x + 0.5)/2.0 - 0.5
            u0 = int(np.floor(u))
            is_ = np.arange(u0-(a-1), u0+a+1); is_c = np.clip(is_, 0, w-1)
            wx = _lanczos_kernel_1d(u - is_, a)
            if normalize and wx.sum() != 0: wx = wx/wx.sum()
            patch = img14[np.ix_(js_c, is_c)]
            out[y, x] = wy @ patch @ wx
    return np.clip(out, 0.0, 1.0)

# -----------------------------
# 2) 4개 방식 업샘플 & 스캔라인 그래프
# -----------------------------
row = 14  # 보고 싶은 행(스캔라인)

SRs = {
    "Nearest":  nearest_x2(LR),
    "Linear":   bilinear_x2(LR),
    "Cubic":    bicubic_x2(LR, a=-0.5),   # Catmull-Rom
    "Lanczos":  lanczos_x2(LR, a=3)       # Lanczos-3
}

x_hr = np.arange(28)
x_lr = np.arange(0, 28, 2)
x_dense = np.linspace(0, 27, 400)

hr_row = HR[row, :]
lr_row = HR[row, ::2]   # 시각화용 샘플 위치(픽셀 스킵 기준)

fig, axes = plt.subplots(2, 2, figsize=(12, 9))
axes = axes.ravel()

for ax, (name, SR) in zip(axes, SRs.items()):
    sr_row = SR[row, :]

    # 시각화(매끈하게 보이도록 1D 보간)
    hr_dense = np.interp(x_dense, x_hr, hr_row)
    sr_dense = np.interp(x_dense, x_hr, sr_row)

    ax.plot(x_dense/27.0, sr_dense, label=f"{name} method")
    ax.plot(x_lr/27.0, lr_row, 'o', mfc='none', mec='orange', label="Sampling points")
    ax.plot(x_dense/27.0, hr_dense, 'k--', label="Exact solution")
    ax.set_title(f"1D interpolation: {name}")
    ax.set_xlabel("Normalized position"); ax.set_ylabel("Intensity")
    ax.set_ylim(0, 1); ax.set_xlim(0, 1)
    ax.legend(loc="lower right")

plt.tight_layout()
plt.show()
```
